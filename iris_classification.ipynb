```python
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import tensorflow as tf
import numpy as np
```

```python
iris = datasets.load_iris()

X = iris.data
X
```

```python
y = iris.target
y
```

```python
scaler_x = StandardScaler()
X = scaler_x.fit_transform(X)
X
```

```python
onehot = OneHotEncoder(categories = 'auto')
y.shape
```

```python
y = y.reshape(-1, 1)
y.shape
```

```python
y = onehot.fit_transform(y).toarray()
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
input_neurons = X.shape[1]
input_neurons
```

```python
hidden_neurons = int(np.ceil((X.shape[1] + y.shape[1]) / 2))
hidden_neurons
```

```python
output_neurons = y.shape[1]
output_neurons
```

```python
W = {'hidden': tf.Variable(tf.random_normal([input_neurons, hidden_neurons])),
     'output': tf.Variable(tf.random_normal([hidden_neurons, output_neurons]))}
```

```python
b = {'hidden': tf.Variable(tf.random_normal([hidden_neurons])),
     'output': tf.Variable(tf.random_normal([output_neurons]))}
```

```python
xph = tf.placeholder('float', [None, input_neurons])
yph = tf.placeholder('float', [None, output_neurons])
```

```python
def mlp(x, W, bias):
    hidden_layer = tf.add(tf.matmul(x, W['hidden']), bias['hidden'])
    hidden_layer_activation = tf.nn.relu(hidden_layer)
    output_layer = tf.add(tf.matmul(hidden_layer_activation, W['output']), b['output'])
    return output_layer
```

```python
model = mlp(xph, W, b)
```

```python
error = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits_v2(logits = model, labels = yph))
optimizer = tf.train.AdamOptimizer(learning_rate = 0.0001).minimize(error)
```

```python
batch_size = 8
batch_total = int(len(X_train) / batch_size)
batch_total
```

```python
X_batches = np.array_split(X_train, batch_total)
X_batches
```

```python
X_batches[1]
```

```python
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    for epoca in range(3000):
        average_error = 0.0
        batch_total = int(len(X_train) / batch_size)
        X_batches = np.array_split(X_train, batch_total)
        y_batches = np.array_split(y_train, batch_total)
        for i in range(batch_total):
            X_batch, y_batch = X_batches[i], y_batches[i]
            _, custo = sess.run([optimizer, error], feed_dict = {xph: X_batch, yph: y_batch})
            average_error += custo / batch_total
        if epoca % 500 == 0:
            print('Época: ' + str((epoca + 1)) + ' error: ' + str(average_error))
    W_final, b_final = sess.run([W, b])
```

```python
W_final
```

```python
b_final
```

```python
predictors_test = mlp(xph, W_final, b_final)
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    r1 = sess.run(predictors_test, feed_dict = {xph: X_test})
    r2 = sess.run(tf.nn.softmax(r1))
    r3 = sess.run(tf.argmax(r2, 1))
```
