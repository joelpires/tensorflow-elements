```python
import pandas as pd
import tensorflow as tf
from sklearn.metrics import mean_absolute_error
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
data = pd.read_csv('house_prices.csv')
data.head()
```

```python
data.columns
```

```python
columns = ['price', 'bedrooms', 'bathrooms', 'sqft_living',
       'sqft_lot', 'floors', 'waterfront', 'view', 'condition', 'grade',
       'sqft_above', 'sqft_basement', 'yr_built', 'yr_renovated', 'zipcode',
       'lat', 'long']
```

```python
columns
```

```python
data = pd.read_csv('house_prices.csv', usecols = columns)
```

```python
data.head()
```

```python
scaler_x = MinMaxScaler()
data[['bedrooms', 'bathrooms', 'sqft_living',
       'sqft_lot', 'floors', 'waterfront', 'view', 'condition', 'grade',
       'sqft_above', 'sqft_basement', 'yr_built', 'yr_renovated', 'zipcode',
       'lat', 'long']] = scaler_x.fit_transform(data[['bedrooms', 'bathrooms', 'sqft_living',
       'sqft_lot', 'floors', 'waterfront', 'view', 'condition', 'grade',
       'sqft_above', 'sqft_basement', 'yr_built', 'yr_renovated', 'zipcode',
       'lat', 'long']])
```

```python
data.head()
```

```python
scaler_y = MinMaxScaler()
data[['price']] = scaler_y.fit_transform(data[['price']])
```

```python
data.head()
```

```python
X = data.drop('price', axis = 1)
y = data.price
```

```python
X.head()
```

```python
type(X)
```

```python
y.head()
```

```python
type(y)
```

```python
predict_columns = columns[1:17]
predict_columns
```

```python
columns_1 = [tf.feature_column.numeric_column(key = c) for c in predict_columns]
```

```python
print(columns_1[10])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
X_test.shape
```

```python
train_function = tf.estimator.inputs.pandas_input_fn(x = X_train, y = y_train,
                                                        batch_size = 32, num_epochs = None, shuffle = True)
```

```python
test_function = tf.estimator.inputs.pandas_input_fn(x = X_test, y = y_test,
                                                   batch_size = 32, num_epochs = 10000, shuffle = False)
```

```python
regressor = tf.estimator.LinearRegressor(feature_columns=columns_1)
```

```python
regressor.train(input_fn=train_function, steps = 10000)
```

```python
train_metrics = regressor.evaluate(input_fn=train_function, steps = 10000)
```

```python
test_metrics = regressor.evaluate(input_fn=test_function, steps = 10000)
```

```python
train_metrics
```

```python
test_metrics
```

```python
prediction_function = tf.estimator.inputs.pandas_input_fn(x = X_test, shuffle = False)
```

```python
predictions = regressor.predict(input_fn=prediction_function)
```

```python
list(predictions)
```

```python
valores_predictions = []
for p in regressor.predict(input_fn=prediction_function):
    valores_predictions.append(p['predictions'])
```

```python
valores_predictions
```

```python
valores_predictions = np.asarray(valores_predictions).reshape(-1,1)
```

```python
valores_predictions
```

```python
valores_predictions = scaler_y.inverse_transform(valores_predictions)
valores_predictions
```

```python
y_test2 = y_test.values.reshape(-1,1)
y_test2.shape
```

```python
y_test2 = scaler_y.inverse_transform(y_test2)
```

```python
y_test2
```
